```python
country=xp.data.gapminder()
lifexp=country.pivot(index="country",columns="year",values="lifeExp")
lifeexp
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(data=lifeexp)
```

```python
Europe=country[country["continent"]=="Europe"].pivot(index="country",columns="year",values="lifeExp")
Europe
```

```python
Europe.sample(10)
```

```python
plt.figure(figsize=(8,10))
sns.heatmap(data=Europe,annot=True,linewidths=0.5,cmap="summer")
```
