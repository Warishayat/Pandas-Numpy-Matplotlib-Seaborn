```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
var=np.random.randint(1,50,50)
```

```python
x=var
x
```

```python
plt.hist(x,color="black",width=5,edgecolor="white",range=(1,50),histtype='barstacked',label="Data")
plt.axvline(43,color="red",label="Red-Zone")
plt.legend()
plt.grid()
plt.title("Histogram",)
plt.xlabel("values")
plt.ylabel('frequency')
plt.show()
```

```python
                                                           #Pie chart
```

```python
y=["Python","Numpy","Pandas","Sk-Learn"]
x=[35,15,35,15]
ex=[0.0,0,0.0,0.2]
autopct_format = '%.2f%%'
plt.pie(x,labels=y,explode=ex,autopct="%0.2f%%",shadow=0.5,labeldistance=1.1,counterclock=False,textprops={"fontsize":10},rotatelabels=False)
plt.legend(loc=2)
plt.title("Pie Chart of language",color="red",)
plt.show()
```
