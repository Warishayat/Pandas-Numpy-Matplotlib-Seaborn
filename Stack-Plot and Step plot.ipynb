```python
x=[1,2,3,4,5]
area=[2,3,4,5,6]
area2=[11,9,1,14,5]
area3=[11,22,23,14,15]
area4=[21,32,3,14,28]
plt.stackplot(x,[area,area2,area3,area4],labels=("area1","area2","area3","area4"),baseline="zero",colors=["black","r","g","b"],)
plt.legend()
plt.xlabel("independent")
plt.ylabel("dependent")
plt.title("STACK-PLOT")
plt.show()
```

```python
x=[21,22,23,24,25]
y=[1,2,3,4,5]
y1=[11,2,3,12,19]
y2=[71,22,1,2,4]
plt.stackplot(x,[y,y1,y2,],colors=["r","g","b"],baseline="wiggle",labels=("y","y1","y2"))
plt.legend()
plt.title("Practise-Stack-plot")
plt.xlabel("x-label")
plt.ylabel("y-label")
plt.show()
```

```python
x=[1,2,3,4,5,6,7]
y=[31,32,42,44,55,7,88]
```

```python
plt.step(x,y,color="red",marker="o",mfc="g",label="python",)
plt.legend(loc=1)
plt.grid()
plt.title("Step-plot")
plt.xlabel("x-var")
plt.ylabel("y-var")
plt.show()
```
