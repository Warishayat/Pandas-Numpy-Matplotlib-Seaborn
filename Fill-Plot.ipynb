```python
x=np.array([1,2,3,4,5,6])
y=np.array([1,2,3,4,5,6])
```

```python
plt.plot(x,y,color="red")
plt.fill_between(x,y,color="yellow",where=(x>=2) & (x<=5),alpha=1)
plt.xlabel("x-values")
plt.ylabel("y-values")
plt.title("Fill-plot")
plt.show()
```
